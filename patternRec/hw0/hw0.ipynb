##Alcohol vs Hue

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'

headers = ['Lab','Alc','Ma','Ash','Alk','Mag','Phe','Fla','NoFla','Pro','Col','Hue','OD','Pr']
out = pd.read_csv('dataset.csv', names=headers)

plt.scatter(out['Alc'], out['Hue'], c=out['Lab'])
plt.xlabel('Alcohol')
plt.ylabel('Hue')
plt.show()
```

The Alcohol vs Hue scatter plot shows that the combination of these two features does not make for a great linear discriminator.
